# 한솥도시락 데이터 수집(메뉴/가격) 

```python
import time
import pandas as pd

from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb

# Keys 모듈 : 키보드의 값들을 가지고 있는 모듈
from selenium.webdriver.common.keys import Keys
```

```python
hsd_url = 'https://www.hsd.co.kr/menu/menu_list'

driver = wb.Chrome()
driver.get(hsd_url)
```

```python
# '더 보기' 버튼을 클릭
for i in range(4):
    driver.find_element_by_css_selector('a.c_05').click()
    time.sleep(1) # 1초 간격으로 클릭
```

# Selenium으로 제어하여 원하는 정보가 웹 사이트에 나왔을 때 bs()를 활용하여 파이썬 객체로 변환!

```python
hsd_html = bs(driver.page_source, 'lxml')
hsd_html
```

### 파이썬 객체로 변환 후 Selenium 종료하기 

```python
driver.close()
```

```python
hsd_names = hsd_html.select('h4.h.fz_03')
hsd_prices = hsd_html.select('span.blind+strong')

len(hsd_names), len(hsd_prices)
```

```python
hsd_nameList = []
hsd_priceList = []

for i in range(len(hsd_names)):
    hsd_nameList.append(hsd_names[i].text)
    hsd_priceList.append(hsd_prices[i].text)   
```

```python
hsd_priceList
```

```python
hsd_dic = {'도시락이름' : hsd_nameList, '가격' : hsd_priceList}
hsd_df = pd.DataFrame(hsd_dic)
```

#### 카테고리별로 상품명, 상품가격 수집하기

```python
gmarket_url = 'http://corners.gmarket.co.kr/BestSellers'

driver = wb.Chrome()
driver.get(gmarket_url)

categorys = driver.find_elements_by_css_selector('#categoryTabG > li > a')

pN_list = []
pP_list = []

for i in range(1, len(categorys)):
    categorys = driver.find_elements_by_css_selector('#categoryTabG > li > a')
    
    categorys[i].click()
    time.sleep(1)
    
    g_html = bs(driver.page_source,'lxml')
    productNms = g_html.select('li.first >a.itemname', limit = 10)
    productPrices =  g_html.select(' div.s-price >strong >span', limit=10)

    for i in range(len(productNms)):
        pN_list.append(productNms[i].text)
        pP_list.append(productPrices[i].text)
```

```python

cate = driver.find_elements_by_css_selector('#categoryTabG > li > a').text
```

```python
print(len(pN_list))
print(len(pP_list))
```

```python
Gmarket_dic = {'카테고리' : categorys , '상품명' : pN_list, '가격' : pP_list}
Gmarket_df = pd.DataFrame(Gmarket_dic)
```

###  유튜브 카테고리별 데이터 수집

```python
youtube_url = 'https://www.youtube.com/'

driver = wb.Chrome()
driver.get(youtube_url)
```

```python
categorys = driver.find_elements_by_css_selector('.style-scope.ytd-feed-filter-chip-bar-renderer > .style-scope.yt-chip-cloud-chip-renderer')

for i in range(len(categorys)):
    categorys = driver.find_elements_by_css_selector('.style-scope.ytd-feed-filter-chip-bar-renderer > .style-scope.yt-chip-cloud-chip-renderer')
    categorys[i].click()
    
    time.sleep(1)
```

```python
len(categorys)
```

```python
y_page = bs(driver.page_source, 'lxml')
title = y_page.select('yt-formatted-string.style-scope.ytd-rich-grid-media', limit = 10)
```

```python
title_list = []

for i in range(len(categorys)):
    categorys = driver.find_elements_by_css_selector('.style-scope.ytd-feed-filter-chip-bar-renderer > .style-scope.yt-chip-cloud-chip-renderer')
    
    categorys[i].click()
    time.sleep(1)
    y_page = bs(driver.page_source, 'lxml')
    title = y_page.select('yt-formatted-string.style-scope.ytd-rich-grid-media', limit = 10)
    
    for i in range(len(title)):
        title_list.append(title[i].text)
```
