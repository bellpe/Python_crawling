#####  차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도

```python
# 1. 차대차 사건 중
# 2. (1) 죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
# 3. 시각화
```

```python
# 모든 컬럼 출력
pd.set_option('display.max_columns',27) # None - 전체 컬럼 출력
```

```python
# 1. 차대차 사건 데이터 가져오기
# 불리언 인덱싱
# (1)사고유형_대분류 컬럼 가져오기
data['사고유형_대분류']
```

```python
# (2) (1)에서 가져온 컬럼으로 '차대차' 데이터 판별
data['사고유형_대분류'] == '차대차' # 불리언
```

```python
# (3) 가져오기
acc =data[data['사고유형_대분류'] == '차대차']
```

```python
acc
```

```python
# 2. (1) 죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
acc['발생지시도'].value_counts()
```

```python
# 사상자수, 발생지 시도 컬럼 가져오기
acc[['사상자수', '발생지시도']]
```

```python
# groupby()
# 같은 값을 묶어(그룹화) 통계 결과를 낼 때 사용
acc_result = acc.loc[:, ['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
x = acc_result.index
x
```

```python
y = acc_result['사상자수'].values
```

```python
# 그래프 시각화
plt.figure(figsize=(10,5)) #(가로, 세로)
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,5)) #(가로, 세로)
plt.bar(x,y)
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.title('차vs차 교통사고의 사상자 수')
plt.show()
```
